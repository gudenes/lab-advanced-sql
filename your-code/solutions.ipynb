```python
from google.cloud import bigquery
```

```python
client = bigquery.Client()
```

```python
query2='''

SELECT 
    Author_ID,
    (Total_Roy_Unique + titles.advance) as Profit
    
FROM (
    SELECT 
        Title_ID,
        Author_ID,
        SUM(Total_Roy) as Total_Roy_Unique
    FROM (
        SELECT
            titles.title_id AS Title_ID,
            auth.au_id AS Author_ID,
            (titles.price * sales.qty * titles.royalty / 100 * titleauth.royaltyper / 100) as Total_Roy

        FROM
            `ironhack-data-analytics.publications.authors` auth

        JOIN 
            `ironhack-data-analytics.publications.titleauthor` titleauth
        ON 
            auth.au_id = titleauth.au_id

        JOIN
            `ironhack-data-analytics.publications.titles` titles
        ON
            titles.title_id = titleauth.title_id

        JOIN   
            `ironhack-data-analytics.publications.sales` sales
        ON
            sales.title_id = titles.title_id

        JOIN   
            `ironhack-data-analytics.publications.roysched` roys
        ON
            sales.title_id = roys.title_id

        GROUP BY
            1,
            2,
            3

    )summary

    GROUP BY 
     1,
     2

    ORDER BY 
     Title_ID
     )summary

JOIN
    `ironhack-data-analytics.publications.titles` titles
ON
    summary.Title_ID = titles.title_id


GROUP BY
1,
2
    
ORDER BY
 Profit DESC

LIMIT 
 3
    


'''
```

```python
query_job = client.query(query2)
df=query_job.to_dataframe()
df
```
